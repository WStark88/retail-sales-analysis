```python
import numpy as np

def cargar_datos(ruta_archivo):
    datos = np.genfromtxt(ruta_archivo, dtype=None, delimiter=',', names=True)
    return datos

if __name__ == "__main__":
    ruta_archivo = '/content/data/retail_sales_dataset.csv'
    data = cargar_datos(ruta_archivo)
    print(data.shape)
    print(data.size)
```

```python
print(data.dtype)
```

```python
print(data.dtype.names)
```

```python
Transaction_ID = np.array([row[0] for row in data])
Date = np.array([row[1] for row in data])
Customer_ID = np.array([row[2] for row in data])
Gender = np.array([row[3] for row in data])
Age = np.array([row[4] for row in data])
Product_Category = np.array([row[5] for row in data])
Quantity = np.array([row[6] for row in data])
Price_Unit = np.array([row[7] for row in data])
Total_Amount = np.array([row[8] for row in data])
```

```python
valores, conteos = np.unique(data['Transaction_ID'], return_counts=True)

# Los duplicados son aquellos cuyo conteo es mayor a 1
duplicados = valores[conteos > 1]

if duplicados.size > 0:
    print(f"Se encontraron {len(duplicados)} IDs duplicados.")
    print(f"IDs repetidos: {duplicados}")
else:
    print("No se encontraron registros duplicados en Transaction_ID. El dataset está limpio.")
```

```python
unique_categories = np.unique(Product_Category)

print("--- Análisis de Ventas por Categoría ---")
for cat in unique_categories:
    mask = (Product_Category == cat)

    total_cat = np.sum(Total_Amount[mask])
    avg_sale = np.mean(Total_Amount[mask])

    print(f"Categoría: {cat}")
    print(f"  Total Ventas: ${total_cat:,.2f}")
    print(f"  Promedio por Transacción: ${avg_sale:,.2f}")
```

este bloque de copdigo calcula el total de ventas por categoría de producto.

```python
totals = np.array([np.sum(Total_Amount[Product_Category == cat]) for cat in unique_categories])

max_idx = np.argmax(totals)
min_idx = np.argmin(totals)

print(f"\nCategoría con MAYOR venta total: {unique_categories[max_idx]} (${totals[max_idx]:,.2f})")
print(f"Categoría con MENOR venta total: {unique_categories[min_idx]} (${totals[min_idx]:,.2f})")
```

```python
cat_search  = 'Electronics'
mask = (Product_Category == cat_search)

total_cat = np.sum(Total_Amount[mask])
avg_sale = np.mean(Total_Amount[mask])

print(f"Resultados para la categoría: {cat_search}")
print(f"Número de registros encontrados: {np.sum(mask)}")
print(f"Suma total de ventas en esta categoría: ${np.sum(avg_sale):,.2f}")
```
