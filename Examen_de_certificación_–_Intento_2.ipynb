### Columnas Nuevas

```python
# Nos seguramos que todos los datos sean del tipo de dato DateTime
# Aplicando la funcion de convertir la columna a datetime (por medida de
# seguridad)
df['release_date'] = pd.to_datetime(df['release_date'])

# Extraer el año y define el tipo de dato como int, permitiendo NaN
df['release_year'] = df['release_date'].dt.year.astype(pd.Int64Dtype())

# Extraer el trimestre y define el tipo de dato como int, permitiendo NaN
df['release_quarter'] = df['release_date'].dt.quarter.astype(pd.Int64Dtype())
```

envista que ya no es imposible imputar las columnas release_year, release_date,release_quarter, el dato NaN sera reemplazado con NaT "Not a Time" lo cual nos evitara promeblas futuros con los graficos.

```python
# Imputamos la FECHA completa primero (usando la fecha conocida en otra consola)
# Usamos 'min' para obtener la fecha de lanzamiento original del título
df['release_date'] = df['release_date'].fillna(
    df.groupby('title')['release_date'].transform('min')
)

# Ahora derivamos el AÑO y el TRIMESTRE de la columna de fecha ya imputada
# Usamos el tipo 'Int64' para que los NaT se mantengan como vacíos y no conviertan todo a decimal
df['release_year'] = df['release_date'].dt.year.astype('Int64')
df['release_quarter'] = df['release_date'].dt.quarter.astype('Int64')

# Verificación de resultados
nulos_finales = df['release_date'].isna().sum()
print(f"Imputación completada.")
print(f"Registros que permanecen como NaT (sin datos en ninguna consola): {nulos_finales}")
```

### Visualizaciones Exploratorias Univariadas

```python
# obtencion valores para la variable categorias
top_cats = df['genre'].value_counts().head(5)

fig, ax = plt.subplots(figsize=(12, 7), facecolor='#242424')

colores = ['#F9596F' if i < 1 else '#A14E9A' if i < 2 else '#FF7A47' for i in range(len(top_cats))]

#carga de datos para el grafico
bars = ax.barh(top_cats.index[::-1], top_cats.values[::-1],
               color=colores[::-1], edgecolor='#0F0F1A', height=0.72)

for bar, val in zip(bars, top_cats.values[::-1]):
    ax.text(bar.get_width() + 8, bar.get_y() + bar.get_height()/2,
            f'{val:,}', va='center', fontsize=13, color='white')
# parametro para la visualizacion del grafico
ax.set_title('Top 5 de las Categorias con mas titulos ', fontsize=14,
             fontweight='bold', color='white')
ax.set_xlabel('Cantidad de Titulos', color='white', fontsize=16)
ax.set_ylabel('Genero', color='white', fontsize=16)
ax.tick_params(axis='x', colors='white', labelsize=9)
ax.tick_params(axis='y', colors='white', labelsize=9)
ax.set_facecolor('#242424')
for spine in ax.spines.values():
    spine.set_color('white')
ax.set_xlim(0, top_cats.max() * 1.15)
ax.grid(axis='x', alpha=0.25)

plt.tight_layout()
plt.show()

print(f'Total de Generos: {df["genre"].nunique()}')
print(f'\nPorcentaje de por Generos:')
for cat, cnt in top_cats.items():
    print(f'  {cat}: {cnt} Titulos ({cnt/len(df)*100:.1f}%)')
```

### Visualizaciones Exploratorias multivariadas

```python
fig, axes = plt.subplots(figsize=(14,7), facecolor='#242424')

fig.suptitle('Distribucion Titulos por Puntuacion ',
             fontsize=14, fontweight='bold', color="white", y=1.01)

# Calculo de Mediana y Media de los mese de permancia.
media   = df['critic_score'].mean()
mediana = df['critic_score'].median()
#carga de datos para el grafico
axes.hist(df['critic_score'], bins=9, color="#FF7A47",
             edgecolor='white', alpha=1, linewidth=1.0)
# parametro para la visualizacion del grafico
axes.axvline(media,   color='#FFA600',  lw=2.5, ls='--', label=f'Media: {media:.2f}')
axes.axvline(mediana, color='white', lw=2.5, ls=':',  label=f'Mediana: {mediana:.2f}')
axes.set_facecolor('#242424')
for spine in axes.spines.values():
    spine.set_color('white')
axes.set_xlabel('¨Puntuacion', color='white')
axes.set_ylabel('Cantidad de Titulos', color='white')
axes.legend(fontsize=10)
axes.grid(axis='y', alpha=0.6, color='white' )
axes.tick_params(axis='x', colors='white')
axes.tick_params(axis='y', colors='white')
plt.tight_layout()
plt.show()

print(f'Media:     {media:.3f}')
print(f'Mediana:   {mediana:.3f}')
print(f'Std:       {df["critic_score"].std():.3f}')
print(f'Asimetria: {df["critic_score"].skew():.3f}  '
      f'({"sesgado a la izquierda" if df["critic_score"].skew() < 0 else "sesgado a la derecha"})')
print(f'Curtosis:  {df["critic_score"].kurtosis():.3f}  '
      f'({"Leptocurtica" if df["critic_score"].kurtosis() < 0 else "Platicurtica"})')
```

* La mayoria de los titulos se encuentran entre 7 y 9 puntos en el ranking, lo cual indica que los titulos  analizados  son principalmente titulos con una recepcion positiva por el mercado.
* Existe una mayor acumulacion de juegos con notas altas.
